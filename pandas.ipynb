```python
def tax(a):
    cont=[]
    for i in a:
        if i<10:
            result=0
        elif 10<=i<20:
            result=i*(5/100)
        elif 20<=i:
            result=i*(30/100)
        cont.append(result)
    return cont 
```

```python
tax(df.Sum)
```

```python
df["Tax"]=tax(df.Sum)
```
